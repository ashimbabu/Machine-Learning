# What is IQR Method?
- Interquartile range
IQR=Q3-Q1
Q1(25th percentile)-value below which 25% of the data falls
Q3(75th percentile)-value below which 75% of the data falls

q1=value at position((n+1)/4)
q3=value at position(3(n+1)/4)
data=[2,4,7,10,12,15,18,21,23]
n=9
calculate positions
q1=(9+1)/4 = 2.5
q3=3(9+1)/4 = 7.5

# interpolate
q1=4+0.5(7-4)=5.5
#q3 is between 7th(18) and 8th(21)
q3=18+0.5(21-18) = 19.5

iqr=q3-q1
=19.5-5.5
=14.5

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

True negative(TN)
    model correctly predicted 0 when actual was 0.
    example: predicted 'fail' and student actual failed
True Positive(TP):--bottom right
    model correctly predicted 1 when actual was 1
    example: predicted 'pass' and students actually passed
False Positive-(top-right):
    model predicted 1, but actual was 0
    example:predicted'pass' , but student actually failed..(Type I error)
False Negative(FN)--Bottom-left:
    model predicted 0, but actual was 1
    example:predicted 'fail', but student actually passed (Type II error)


```python
df['loan_status'].value_counts()
```

```python
#accuracy=(TP+TN)/(TN+TP+FN+FP)
```

```python
#Precision
-out of all predicted positives, how many are actually positive?
precision=TP/(TP+FP)
```

```python
prec=precision_score(y_test,y_pred)
prec
```

```python
# RecaLL(sensitivity/true positive rate)
recall=(TP)/TP+FN
-how many actual positives were correctly predicated
```

```python
rec=recall_score(y_test,y_pred)
rec
```

```python
f1=f1_score(y_test,y_pred)
f1
```

f1_score
-harmonic mean if precision and recall
f1=2.(precision*recall)/(precision+recall)
