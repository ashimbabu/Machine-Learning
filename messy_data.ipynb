# Regex
- a regular expressions is a sequence of characters that defines a search pattern.
    - finding strings
    - replacing text
    - validating inputs

# basic regex syntax (regex, patterens)
Pattern                       matches
.                             any character except newline
^                             start of a string
$                             end of string
*                             0 or more repetations
+                             1 or more repetitations
?                             0 or 1 repitation
[]                            any characters inside a brackets
[^]                           not the characters inside
{n}                           exactly n repetitions
{n,}                          n or more repetitions
{n,m}                         between n and m repetitions

# Common Pattern
\d                            digit(0-9)
\D                            non-digit
\w                            wordcharacter(a-z,A-Z,0-9,_)
\W                            non_word character
\s                            Whitespace
\S                            non_white space
    

```python
import re
```

```python
text='apple'
```

```python
match=re.search(r'a....',text)
if match:
    print('matched')
else:
    print('no match')
```

```python
# ^ - start of a string
text='apple 123 456'
text1='ball'
re.findall(r'^a',text)
```

```python
# ^ - start of a string - $
text='apple 123 456'
text1='ball'
re.findall(r'456$',text)
```

```python
# * 0 or more repetations
text='apple apple pple'
re.findall(r'p*',text)
```

```python
text='apple apple pple'
re.search(r'p*',text)
```

```python
# [] match any character inside
text='apple'
print(re.findall(r'[aeiou]',text))
```

```python
# [] match any character not inside
text='appleaeiou'
print(re.findall(r'[^aeiou]',text))
```

```python
# {n} exactly n repetitions
text='apppple'
re.findall(r'p{3}',text)
```

```python
# {n,m} between n and m
text='appppppppple'
re.findall(r'p{2,5}',text)
```

# Basic Functions in re
1. re.match(pattern,string)
-check if the pattern matchs from the begining of the string

```python
text='python is fun'
print(re.match('python',text))
print(re.match('fun',text))
```

# 2. re.search(pattern,string)
-searches the entire string for the first match

```python
print(re.search('fun',text))
```

# 3. re.findall(pattern,string)
- return all matches as a list

```python
print(re.findall('fun',text))
```

# 4. re.sub(pattern,replace,string)
-replace all matches with something else

```python
text='i love Java.Java is great'
print(re.sub('Java','',text))
```

```python
print(re.sub(' ','',text))
```

# 5.re.split(pattern,string)
-split a string a regex pattern

```python
text='one,two;three,four'
print(re.split('[,;]',text))
```
