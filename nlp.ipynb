```python
# What is Natural Language Processing?
- NLP is the technology that is used by machines to understand, analyses, manipulate, and intrepret human's languages. 
# Major Steps in NLP
1. Text Acquisition:
collecting data from documents, web pages, chat, social media etc.

2. Text cleaning/preprocessing
    - Tokenization             spliting text into words or sentences 'analysis for ML' --> ['analysis','for','ML']
    - lowercasing
    - Removing stopwards       removing unimportant words(a,an,the,is,etc)
    - stemming                 reduce word to its root form                   running, runs --- run
    - lemmatization            convert to dictionary root form (accurate)     better --- good
    removing puncutation/numbers      clean text for analysis

3. Text representation:
    Method:
    bag of words(BoW)          count how many times each word appears
    TF-IDF                     weighted frequency (gives importance to rare words)   used in document classification
    word embeddings            words are represented as dense numberic vectors capturing meaning word2Vec, GloVe, BERT embedding

4. Model Building/NLP Tasks:
    text classification        categorize text       spam or not, sentiment anaysis 
    .....
        
5. Evaluation
    Metric                     Meaning                            Example
    accuracy                   % pf correct prediction            spam detection accuracy
    precision/recall/f1        measure quality of classification  sentiment detection
    BLEU Score                 translation accuracy               machine translation models
    ROUGE Score                summarization quality              text summarizer

6. Deployment 
    
```
