```python
# Union
set1={'a','b','c'}
set2={1,2,3}

set3=set1.union(set2)
print(set3)
```

```python
# Union
set1={'a','b','c'}
set2={1,2,3}
set4={6,7,90}

set3=set1 | (set2) | set4
print(set3)
```

```python
# intersection 
set1={'apple','banana',1,2,3,4,5,6,7}
set2={'apple','banana',3,4,5,6,7,8,9,0}
set3=set1 & set2
print(set3)
```

```python
# difference
set1={'apple','banana',1,2,3,4,5,6,7}
set2={'google','apple',3,4,5,6,7,8,9,0}
set3=set2.difference(set1)
print(set3)
```

```python
# 2. Solutions
class Shape:
    def __init__(self, length=0, width=0, side=0, base=0, height=0):
        self.length = length
        self.width = width
        self.side = side
        self.base = base
        self.height = height

    def area_rectangle(self):
        return self.length * self.width

    def area_square(self):
        return self.side * self.side

    def area_triangle(self):
        return 0.5 * self.base * self.height

shape = Shape(length=10, width=5, side=4, base=6, height=3)

print("Area of Rectangle:", shape.area_rectangle())  # 10 * 5 = 50
print("Area of Square:", shape.area_square())        # 4 * 4 = 16
print("Area of Triangle:", shape.area_triangle())    # 0.5 * 6 * 3 = 9
```

```python
# Inheritance
- Inheritance allows us to define class that inherits all methods and properties from another class.

parent class is the class being inherited from, also called base class
child class is the class that inherits from another class, also called derived class

1. Single Inheritance
```

```python
# 1. Single Inheritance
class Animal:
    def make_sound(self):
        print('every animal breathe')
        
class Dog(Animal):
    def bark(self):
        print('the dog bark woof woof')
        
# always create the object of child class
dog=Dog()
dog.bark()
dog.make_sound()
```

```python
class Vehicle:
    def __init__(self,brand: str,speed: int):
        self.brand: str=brand
        self.speed: int=speed

    def move(self):
        return f"{self.brand} is moving at speed of {self.speed} km/h"
class Car(Vehicle):
    def __init__(self,brand:str,speed: str,num_doors=int):
        super().__init__(brand,speed)
        self.num_doors:int=num_doors

    def move(self):
        return f"{self.brand} is moving at speed of {self.speed} km/h with {self.num_doors} doors"
car=Car(brand='Toyota',speed=120,num_doors=4)
car.move()
```

```python
# Single inheritance
# Parent class
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def show_details(self):
        print(f"Name: {self.name}, Salary: {self.salary}")

# Child class
class Manager(Employee):
    def assign_task(self, task):
        print(f"{self.name} assigns task: {task}")

# Example usage
mgr = Manager('Ram', 75000)
mgr.show_details()    # Inherited from Employee
mgr.assign_task("Prepare Survey Report")  # From Manager
```

```python
# Multilevel Inheritance
# Grandparent class
class Gadget:
    def __init__(self, brand):
        self.brand = brand
    
    def turn_on(self):
        print(f'{self.brand} gadget is now ON.')

# Parent class
class Phone(Gadget):
    def call(self, number):
        print(f'Calling {number} from {self.brand} phone.')

# Child class
class Smartphone(Phone):
    def browse_internet(self):
        print(f'{self.brand} smartphone is browsing the internet.')

# Example usage
my_phone = Smartphone('Samsung')
my_phone.turn_on()       # From Gadget
my_phone.call("98510-45361") # From Phone
my_phone.browse_internet() # From Smartphone
```

```python
class ElectricVehicle:
    def __init__(self,battery_capacity: int):
        self.battery_capacity: int=battery_capacity

    def charge(self) -> str:
        return f"charging battery to {self.battery_capacity} kwh"

class FlyingMachine:
    def fly(self) -> str:
        return 'Flying in the sky'

class ElectricFlyingCar(ElectricVehicle, FlyingMachine):

    def __init__(self,brand: str,battery_capacity:int):
        ElectricVehicle.__init__(self,battery_capacity)
        self.brand: str=brand

    def move(self) -> str:
        return f"{self.brand} is driving and flying with {self.battery_capacity} kwh"
car=ElectricFlyingCar('skydrive',75)
print(car.move())
print(car.fly())
print(car.charge())
```

```python
class Device:
    def __init__(self,brand:str):
        self.brand:str=brand
    def power_on(self) -> str:
        return f"{self.brand} device is now on"

class Smartphone(Device):
    def call(self) -> str:
        return f"{self.brand} smartphone is making a call"
        
class Laptop(Device):
    def code(self) -> str:
        return f"{self.brand} laptop is making a code"

class Tablet(Device):
    def draw(self) -> str:
        return f"{self.brand} tablet is used for drawing"

# object of class 1 smartphone
phone=Smartphone('Samsung')
print(phone.call())
print(phone.power_on())

lap=Laptop('Dell')
print(lap.code())
print(lap.power_on())

tab=Tablet('Apple')
print(tab.draw())
print(tab.power_on())
```

```python
# Lets create a structure
- Vehicle -> Base class
- electric -> adds electric functionality
- car -> inherits from vehicle 
- electricCar - inherits from both car and electric 
```

```python
# Base Class
class Person:
    def __init__(self, name):
        self.name = name
    
    def show_name(self):
        print(f"Name: {self.name}")

# First Level of Inheritance
class Employee(Person):
    def __init__(self, name, salary):
        super().__init__(name)
        self.salary = salary
    
    def show_salary(self):
        print(f"Salary: {self.salary}")

class Student(Person):
    def __init__(self, name, course):
        super().__init__(name)
        self.course = course
    
    def show_course(self):
        print(f"Course: {self.course}")

# Hybrid: Multiple Inheritance
class Intern(Employee, Student):
    def __init__(self, name, salary, course, duration):
        Employee.__init__(self, name, salary)
        Student.__init__(self, name, course)
        self.duration = duration
    
    def show_internship(self):
        print(f"Internship Duration: {self.duration} months")

# Example usage
intern1 = Intern('Ailesh', 20000, 'Information Technology', 6)
intern1.show_name()        # From Person
intern1.show_salary()      # From Employee
intern1.show_course()      # From Student
intern1.show_internship()  # From Intern
```
